```python
# 1) Find the student name who scored maximum scores in all (exam, quiz and homework)?
```

```python
# Exam - max mark
max_exam=collection.aggregate([{"$unwind":"$scores"}, {"$match":{"scores.type":"exam"}}, {"$group":{"_id":"null","maxi":{"$max":"$scores.score"}}}])
for i in max_exam:
    maxi=i["maxi"]
exam=collection.find({"scores.score":maxi},{"_id":1,"name":1})
for i in exam:
    Name=i["name"]
    print(f'Maximum mark in exam of {maxi} was scored by {Name}.')
    
# Quiz - max mark
max_quiz=collection.aggregate([{"$unwind":"$scores"}, {"$match":{"scores.type":"quiz"}}, {"$group":{"_id":"null","maxi":{"$max":"$scores.score"}}}])
for i in max_quiz:
    maxi=i["maxi"]
quiz=collection.find({"scores.score":maxi},{"_id":1,"name":1})
for i in quiz:
    Name=i["name"]
    print(f'Maximum mark in quiz of {maxi} was scored by {Name}.')

# Homework - max mark
max_homework=collection.aggregate([{"$unwind":"$scores"}, {"$match":{"scores.type":"homework"}}, {"$group":{"_id":"null","maxi":{"$max":"$scores.score"}}}])
for i in max_homework:
    maxi=i["maxi"]
quiz=collection.find({"scores.score":maxi},{"_id":1,"name":1})
for i in quiz:
    Name=i["name"]
    print(f'Maximum mark in homework of {maxi} was scored by {Name}.')
```

```python
# 2) Find students who scored below average in the exam and pass mark is 40%?
```

```python
avgMarks=collection.aggregate([{"$unwind":"$scores"},{"$match":{"scores.type":"exam"}},{"$group":{"_id":"null","avgMark":{"$avg":"$scores.score"}}}])
for i in avgMarks:
  avgMark=i["avgMark"]
belowAvg=collection.aggregate([{"$unwind":"$scores"},{"$match":{"scores.type":"exam"}},{"$match":{"scores.score":{"$lt":avgMark}}}])
for i in belowAvg:
  print(i['name'])
```

```python
# 3) Find students who scored below pass mark and assign them as fail, and above pass mark as pass in all the categories.
```

```python
# Exam
result=collection.aggregate([{"$unwind":"$scores"},{"$match":{"scores.type":"exam"}},{"$match":{"scores.score":{"$lt":40}}}])
for i in result:
  collection.update_one({"_id":i["_id"]},{"$push":{"result.exam_result":"Fail"}})

result=collection.aggregate([{"$unwind":"$scores"},{"$match":{"scores.type":"exam"}},{"$match":{"scores.score":{"$gte":40}}}])
for i in result:
  collection.update_one({"_id":i["_id"]},{"$push":{"result.exam_result":"Pass"}})

# Quiz
result=collection.aggregate([{"$unwind":"$scores"},{"$match":{"scores.type":"quiz"}},{"$match":{"scores.score":{"$lt":40}}}])
for i in result:
  collection.update_one({"_id":i["_id"]},{"$push":{"result.quiz_result":"Fail"}})

result=collection.aggregate([{"$unwind":"$scores"},{"$match":{"scores.type":"quiz"}},{"$match":{"scores.score":{"$gte":40}}}])
for i in result:
  collection.update_one({"_id":i["_id"]},{"$push":{"result.quiz_result":"Pass"}})

# Homework
result=collection.aggregate([{"$unwind":"$scores"},{"$match":{"scores.type":"homework"}},{"$match":{"scores.score":{"$lt":40}}}])
for i in result:
  collection.update_one({"_id":i["_id"]},{"$push":{"result.homework_result":"Fail"}})

result=collection.aggregate([{"$unwind":"$scores"},{"$match":{"scores.type":"homework"}},{"$match":{"scores.score":{"$gte":40}}}])
for i in result:
  collection.update_one({"_id":i["_id"]},{"$push":{"result.homework_result":"Pass"}})

for i in collection.find():
    print(i)
```

```python
# 4) Find the total and average of the exam, quiz and homework and store them in a separate collection.
```

```python
new_collection=db.Avg_Total
```

```python
# Exam - Total and Average
exam_total=collection.aggregate([{"$unwind":"$scores"},{"$match":{"scores.type":"exam"}},{"$group":{"_id":"null","exam_total":{"$sum":"$scores.score"}}}])
for i in exam_total:
  exam_total=i["exam_total"]
exam_avg=collection.aggregate([{"$unwind":"$scores"},{"$match":{"scores.type":"exam"}},{"$group":{"_id":"null","exam_avg":{"$avg":"$scores.score"}}}])
for i in exam_avg:
  exam_avg=i["exam_avg"]

# Quiz - Total and Average
quiz_total=collection.aggregate([{"$unwind":"$scores"},{"$match":{"scores.type":"quiz"}},{"$group":{"_id":"null","quiz_total":{"$sum":"$scores.score"}}}])
for i in quiz_total:
  quiz_total=i["quiz_total"]
quiz_avg=collection.aggregate([{"$unwind":"$scores"},{"$match":{"scores.type":"quiz"}},{"$group":{"_id":"null","quiz_avg":{"$avg":"$scores.score"}}}])
for i in quiz_avg:
  quiz_avg=i["quiz_avg"]

# Homework - Total and Average
homework_total=collection.aggregate([{"$unwind":"$scores"},{"$match":{"scores.type":"homework"}},{"$group":{"_id":"null","homework_total":{"$sum":"$scores.score"}}}])
for i in homework_total:
  homework_total=i["homework_total"]
homework_avg=collection.aggregate([{"$unwind":"$scores"},{"$match":{"scores.type":"homework"}},{"$group":{"_id":"null","homework_avg":{"$avg":"$scores.score"}}}])
for i in homework_avg:
  homework_avg=i["homework_avg"]

results=[{'Type':"Exam", "Total":exam_total, "Average":exam_avg},{'Type':"Quiz", "Total":quiz_total, "Average":quiz_avg},{'Type':"Homework", "Total":homework_total, "Average":homework_avg}]
new_collection.insert_many(results)
```

```python
for i in new_collection.find():
    print(i)
```

```python
# 5) Create a new collection which consists of students who scored below average and above 40% in all the categories.
```

```python
belowAvg_above40=db.belowAvg_above40
```

```python
# Exam
below_above=collection.aggregate([{"$unwind":"$scores"},{"$match":{"scores.type":"exam"}},{"$match":{"scores.score":{"$gt":40,"$lt":exam_avg}}}])
for i in below_above:
  Name=i['name']
  Type=i['scores']['type']
  belowAvg_above40.insert_one({'Name':Name,'Type':Type})

# Quiz
below_above=collection.aggregate([{"$unwind":"$scores"},{"$match":{"scores.type":"quiz"}},{"$match":{"scores.score":{"$gt":40,"$lt":quiz_avg}}}])
for i in below_above:
  Name=i['name']
  Type=i['scores']['type']
  belowAvg_above40.insert_one({'Name':Name,'Type':Type})

# Homework
below_above=collection.aggregate([{"$unwind":"$scores"},{"$match":{"scores.type":"homework"}},{"$match":{"scores.score":{"$gt":40,"$lt":homework_avg}}}])
for i in below_above:
  Name=i['name']
  Type=i['scores']['type']
  belowAvg_above40.insert_one({'Name':Name,'Type':Type})
```

```python
for i in belowAvg_above40.find():
    print(i)
```

```python
# 6) Create a new collection which consists of students who scored below the fail mark in all the categories.
```

```python
fail_studs=db.failed_students
```

```python
fail=collection.aggregate([{"$unwind":"$result"},{"$match":{"result.exam_result":"Fail"}},
                        {"$match":{"result.quiz_result":"Fail"}},{"$match":{"result.homework_result":"Fail"}}])
for i in fail:
  Name=i['name']
  fail_studs.insert_one({'Name':Name})
```

```python
for i in fail_studs.find():
    print(i)
```

```python
# 7) Create a new collection which consists of students who scored above pass mark in all the categories.
```

```python
pass_studs=db.passed_students
```

```python
passed=collection.aggregate([{"$unwind":"$result"},{"$match":{"result.exam_result":"Pass"}},
                        {"$match":{"result.quiz_result":"Pass"}},{"$match":{"result.homework_result":"Pass"}}])
for i in passed:
  Name=i['name']
  pass_studs.insert_one({'Name':Name})
```

```python
for i in pass_studs.find():
    print(i)
```
